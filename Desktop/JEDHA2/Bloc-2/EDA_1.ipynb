```python
#Compter le nombre de lignes par ville sur la colonne 'from'
df_city = df_raw['from'].value_counts().reset_index(drop=False).rename(columns={'index':'City','from': 'Nb'})

df_city.head()
```

```python
#Ajouter la latitude et longitude de chaque Ville au dataframe 'df_city'

idx = 0
for i in df_city['City']:
    if i == i: #Permet d'ignorer les Nan
        key = "2439967aa2c04b77bae47c32a7549f95"
        geocoder = OpenCageGeocode(key)
        query = i
        results = geocoder.geocode(query)
        df_city.loc[idx,"latitude"] = results[0]['geometry']['lat']
        df_city.loc[idx,"longitude"] = results[0]['geometry']['lng']

        idx += 1
    else :
        pass
    
df_city.head(5)
```

```python
#Export du dataframe en CSV
df_city.to_csv('lat_long.csv')
```

```python
#Joindre les 2 dataframe sur les colonnes "City"
df_raw = df_raw.rename(columns={'from':'City'}) #On doit d'abord renommer pour avoir le mÃªme nom de colonne
df_raw2 = df_raw.merge(df_city, on='City')
```

```python
#Affichage des 3 derniÃ¨res colonnes du nouveau dataset -> 'on voit bien que les colonnes ont Ã©tÃ© ajoutÃ©es'
df_raw2.columns.tolist()[-3:]
```

```python
#Origine des participants
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = 'iframe_connected'

fig = go.Figure(data=go.Scattergeo(
        lon = df_raw2['longitude'],
        lat = df_raw2['latitude'],
        text = df_raw2['Nb'],
        mode = 'markers',
        marker_color = df_raw2['Nb'],
        ))

fig.update_layout(
        title = 'Origine des participants'
    )
fig.show()
```

```python
#Ville d'origine des participants -> Plus le point est gros plus ils sont nombreux Ã  Ãªtre originaires de la ville

import plotly.graph_objects as go
import plotly.express as px


fig = px.scatter_mapbox(df_raw2, 
                        lat='latitude', 
                        lon='longitude', 
                        zoom=1, 
                        color = 'City',
                        size = 'Nb',
                        mapbox_style="carto-positron", 
                        title = "Ville d'origine des participants")
fig.show()

#Les Figurants sont principalement originaires des US ou d'Europe Occidentale avec quelques argentins et sud-asiatiques, on a Ã©galement un grand nombre d'israÃ©liens.
# En revanche pas d'Est-EuropÃ©ens ni d'Africains ni de Nord-Asiatiques ou Australiens
```

```python
#Importance of the race according to the homecountry they from 

import plotly.express as px
import matplotlib.pyplot as plt

fig = px.scatter_mapbox(df_raw2, lat="latitude", lon="longitude", color="imprace", mapbox_style="carto-positron", color_continuous_scale=px.colors.diverging.RdBu[::-1] , zoom=0)
fig.show()

#We can see that for instance Chinese, Brazilians, Israelians and even some Americans (Iowa..) consider Race as very Important
```
